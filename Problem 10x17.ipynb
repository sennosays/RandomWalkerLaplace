```julia
nRows = 30; 
nCols = 25; 
size = (nRows, nCols); 
grid = zeros(Float64,size);

#set boundary conditions 
grid[1,:] = 1.0;       #top 
grid[nRows,:] = 2.0;  #bottom
grid[:,1] = 3.0;       #left 
grid[:,nCols] = 4.0;   #right 

original_grid = copy(grid);  
```

```julia
function grid_solve(my_grid::Array,tol = 0.0001::Float64)
    new_grid = copy(my_grid);
    max_diff = 1.0; 
    max_iterations = 10000; 

    for i in 1:max_iterations
        while(max_diff > tol)
            max_diff = 0.0; 
            for j in 2:nRows-1
                for k in 2:nCols-1
                    old_value = my_grid[j,k]; 
                    new_value = 0.25*(my_grid[j+1,k]+my_grid[j-1,k]+my_grid[j,k+1]+my_grid[j,k-1]);
                    this_diff = abs(new_value-old_value); 
                    if(this_diff>max_diff)
                        max_diff = this_diff;
                    end
                    new_grid[j,k] = new_value;
                end
            end
            my_grid = copy(new_grid); 
        end
    end
    return my_grid
end
```

```julia
grid=grid_solve(grid);
```

```julia
function random_walker(nRows::Int,nCols::Int,initial_index::Array{Int64,2})
    next = [-1 0; 0 1; 1 0; 0 -1];
    index = initial_index; 
    done = index_at_boundary(initial_index,nRows,nCols); 
    while(!done)
        get_next_index = rand(1:4);
        index += next[get_next_index,:]; 
        done = index_at_boundary(index,nRows,nCols); 
    end
    return index
end
```

```julia
#function that determines if the walker has reached the boundary 
#bottom should be the number of rows for the grid
#right should be the number of coloumns for the grid 

function index_at_boundary(my_index::Array{Int64,2},bottom::Int,right::Int)
    return (my_index[1] == 1 || my_index[1]==bottom || my_index[2] == 1 || my_index[2] == right)
end
```

```julia
function random_walk_grid_solver(grid::Array, starting_index::Array{Int64,2},num_walkers= 1000::Int)
    this_nRows = length(grid[:,1]); 
    this_nCols = length(grid[1,:]); 
    
    potential = 0.0; 
    
    
    
    for i in 1:num_walkers
        ending_index = random_walker(this_nRows,this_nCols,starting_index);
        potential += grid[ending_index[1],ending_index[2]]; 
    end
    
    return potential/num_walkers; 


end
```
